```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('../data/Presentation_df',index_col="Id")
coef_df = pd.read_csv('../data/coef_df',index_col=0)
base_data = pd.read_csv('../data/train.csv')
```

```python
df["Residual"] = df["Predictions"] - df["SalePrice"]
```

```python
f,ax = plt.subplots(figsize=(15,8))
sns.regplot(df["SalePrice"],df['Predictions'],ax=ax)
```

```python
top_coefs = coef_df.sort_values(by='Coefficient',ascending=False).head(10)
```

```python
bottom_coefs = coef_df.sort_values(by='Coefficient',ascending=True).head(10)
```

```python
f, ax = plt.subplots(figsize=(18,8))
sns.barplot(x=top_coefs.index,y=top_coefs["Coefficient"],ax=ax)
```

```python
plt.figure(figsize=(15,8))
sns.pointplot(x=base_data["Bsmt Qual"],y=base_data['SalePrice'],order=['Fa','TA','Gd', 'Ex'],orient='v',ci=0)
plt.xlabel("Basement Quality",fontsize = 20)
plt.ylabel("Sale Price",fontsize = 20)
```

```python
plt.figure(figsize=(20,10))
sns.pointplot(x=base_data["Kitchen Qual"],y=base_data['SalePrice'],order=['Fa','TA','Gd', 'Ex'],orient='v',ci=0)
plt.xlabel("Kitchen Quality",fontsize = 20)
plt.ylabel("Sale Price",fontsize = 20)
```
